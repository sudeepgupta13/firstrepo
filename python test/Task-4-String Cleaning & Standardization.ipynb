```python
import pandas as pd
import numpy as np
df = pd.read_csv('movies.csv')
print(df.columns)
str_cols = df.select_dtypes(include='object').columns
df[str_cols] = df[str_cols].apply(lambda col: col.str.strip())
df
```

```python
str_cols = df.select_dtypes(include='object').columns
df[str_cols] = df[str_cols].apply(lambda col: col.str.lower())
df
```

```python

df['movie_name'] = df['movie_name'].str.replace(r"['\".]", "", regex=True)
df['year_of_release'] = df['year_of_release'].str.replace(r"['\".]", "", regex=True)
df['genre'] = df['genre'].str.replace(r"['\".]", "", regex=True)
df
```
